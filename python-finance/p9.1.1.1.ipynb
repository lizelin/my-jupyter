```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def f(x):
    return np.sin(x) + 0.5*x
x = np.linspace(-2 * np.pi, 2*np.pi, 50)
plt.plot(x, f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
plt.show()
```

```python
reg = np.polyfit(x, f(x), deg=1)
ry = np.polyval(reg, x)

```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')

```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
reg = np.polyfit(x, f(x), deg=7)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# 检查是否拟合完美
np.allclose(f(x), ry)
```

```python
# 看均方差（MSE）区间是否较窄
np.sum((f(x)-ry)**2)/len(x)
```

```python
matrix = np.zeros((3+1, len(x)))
matrix[3,:] = x ** 3
matrix[2,:] = x ** 2
matrix[1,:] = x
matrix[0,:] = 1

```

```python
reg1 = np.linalg.lstsq(matrix.T, f(x))[0]
reg1
```

```python
ry1 = np.dot(reg1, matrix)
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry1, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
matrix[3:] = np.sin(x)
reg2 = np.linalg.lstsq(matrix.T, f(x))[0]
ry2 = np.dot(reg2, matrix)
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry2, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
np.allclose(f(x), ry2)

```
